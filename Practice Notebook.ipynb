# Creating a simple plot - Notice the pip installs at the top 

```python
%matplotlib inline
```

```python
"""
Creating a plot in line - you need to call the above "special function" to enable inline plots 
lsmagic
"""

import numpy as np
import matplotlib.pyplot as plt

N = 50
x= np.random.rand(N)
y= np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 # ) to 15 point radius

plt.scatter(x,y, s= area, c= colors, alpha = 0.5)
plt.show()
```
